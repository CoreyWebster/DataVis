Code for 5 random lines
This program simply takes 10 random points, 5(x,y) start and 5(x,y) stop and draws lines between each start and stop point at random. 

```python
from random import *
import matplotlib.pyplot as plt
import pylab


x1_start = randint(1, 768)
x1_end = randint(1, 768)
y1_start = randint(1, 768)
y1_end = randint(1, 768)

x2_start = randint(1, 768)
x2_end = randint(1, 768)
y2_start = randint(1, 768)
y2_end = randint(1, 768)

x3_start = randint(1, 768)
x3_end = randint(1, 768)
y3_start = randint(1, 768)
y3_end = randint(1, 768)

x4_start = randint(1, 768)
x4_end = randint(1, 768)
y4_start = randint(1, 768)
y4_end = randint(1, 768)

x5_start = randint(1, 768)
x5_end = randint(1, 768)
y5_start = randint(1, 768)
y5_end = randint(1, 768)

plt.plot([x1_start,y1_start],[y1_start,y1_end] , 'black')
plt.plot([x2_start,y2_start],[y2_start,y2_end] , 'black')
plt.plot([x3_start,y3_start],[y3_start,y3_end] , 'black')
plt.plot([x4_start,y4_start],[y4_start,y4_end] , 'black')
plt.plot([x5_start,y5_start],[y5_start,y5_end] , 'black')
plt.ylim(0,768)
plt.xlim(0,768)

fig = plt.show()
```
